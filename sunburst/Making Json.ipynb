```python
nest = dict()
nest = {'name':'flare','data':[]}

for line in df.values:
    flare = line[0]
    Types = line[1]
    Colour_Code	 = line[2]
    Type = line[3]
    value = line[4]
    
    flare_list = []
    for item in nest['data']:
        flare_list.append(item['flare'])
    
    if not flare in flare_list:
        nest['data'].append({
            'flare':flare,
            'data':[{
                'name':Types,
                'colour':Colour_Code,
                'data':[{
                    'name':Type,'value':value
                }]
            }]
        })
        
    else:
        nest['data'][flare_list.index(flare)]['data'].append({
            'flare':flare,
            'data':[{
                'name':Types,
                'colour':Colour_Code,
                'data':[{
                    'name':Type,'value':value
                }]
            }]
        })
```

```python
def allflare(df, flare,filename):
    flare = dict()
    flare = {'flare':flare,'children':[]}
    
    for line in df.values:
        Types = line[0]
        Colour_Code = line[1]
        Type = line[2]
        value = line[3]
        
        keys_list=[]
        for item in flare['children']:
            keys_list.append(item['name'])
        if not Types in keys_list:
            flare['children'].append({
                'name':Types,
                'children':[{'name':Type,'value':value}]
            })
        else:
            flare['children'][keys_list.index(Types)]['children'].append({
                'name':Type,
                'value':value
            })
    #return flare
    with open(filename,'w') as outfile:
        json.dump(flare,outfile)
```

```python
allflare(df_m,flare,'flarejsonall.json')
```
